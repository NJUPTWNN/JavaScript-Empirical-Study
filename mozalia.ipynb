```python
import pandas as pd
```

```python
df = pd.read_excel("spidermonkey-pr.xlsx")
df.head()                   
```

```python
df = df[df['修改文件路径'].notnull()]
df.head()
```

```python
df.info()
```

```python
## 最初的file列表
file_list = df['修改文件路径'].to_list()
len(file_list)
```

```python
file_list
```

```python
from collections import Counter
Counter(file_list)
```

```python
dict_complete = dict(Counter(file_list))
```

```python
df_c = pd.DataFrame(dict_complete,index=[1]).T
df_c.head()
```

```python
df_c.to_csv("chakra_complete.csv")
```
