```python
data = df[600:900]
url_list = data['url'].to_list()
print(url_list)
# url_list = df['url'].to_list()
# len(url_list)
# print(url_list)
```

```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
num = 0
for item_url in url_list:
    num += 1
    print("正在处理第"+str(num)+"个")
    results_list = [item_url]
    chrome_options = Options()
    chrome_options.add_argument('--headless')
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(item_url)
    # driver.maximize_window()
    sleep(5)
    try:
        # wait = WebDriverWait(driver, 20)
        ele1 = driver.find_element(By.ID,"pg-app")
        res1 = driver.execute_script("return arguments[0].shadowRoot",ele1)

        ele2 = res1.find_element(By.ID,"app-element")
        res2 = driver.execute_script("return arguments[0].shadowRoot",ele2)

        ele3 = res2.find_element(By.CSS_SELECTOR,"main > gr-change-view")
        res3 = driver.execute_script("return arguments[0].shadowRoot",ele3)

        ID = res3.find_element(By.CSS_SELECTOR, "#mainContent > section > div > div > a").text
        results_list.append(ID)

        ele4 = res3.find_element(By.CSS_SELECTOR,"#commitMessageEditor > gr-linked-text")
        res4 = driver.execute_script("return arguments[0].shadowRoot",ele4)

        # cont = res4.find_element(By.CSS_SELECTOR,"#output").text
        # results_list.append(cont)

        ele5 = res3.find_element(By.CSS_SELECTOR,"#fileList")
        res5 = driver.execute_script("return arguments[0].shadowRoot",ele5)

        stickyArea_list = res5.find_elements(By.CSS_SELECTOR,"#container > div.stickyArea")
        if stickyArea_list:
            del(stickyArea_list[0])
            # pathList = []
            add_list = []
            rem_list = []
            result_list = []
            for item in stickyArea_list:
                #  div > span > a > span.fullFileName
                # path = stickyArea_list[1].find_element(By.CSS_SELECTOR,"div > span > a > span.fullFileName").text
                # path = item.find_element(By.CSS_SELECTOR,"div > span > a > span.fullFileName").text
                add_num = item.find_element(By.CSS_SELECTOR,"div > div.stats > div > span.added").text
                rem_num = item.find_element(By.CSS_SELECTOR,"div > div.stats > div > span.removed").text
                add_list.append(int(add_num))
                rem_list.append(abs(int(rem_num)))
                # pathList.append(path)
                result_list = [add_list[i] + rem_list[i] for i in range(len(add_list))]

            count = len(result_list)
            results_list.append(count)
            results_list.append(result_list)
            # results_list.append(pathList)
            driver.close()
            csvfile = open('set_result.csv','a+',newline='')
            writer = csv.writer(csvfile)
            writer.writerow(results_list)
            print("succ ")
        else:
            print("没有修改文件")
            continue
    except NoSuchElementException:
        continue
    except AttributeError:
        continue

csvfile.close()
```
